 ### Arrays, Matrizen und elementare Lineare Algebra

Dieses Kapitel rekapituliert die wichtigsten Elemente der Linearen Algebra, die im Bereich des Maschinellen Lernens Anwendung finden. 

---

-  Eine Matrix $\mathbf{A}\in \mathbb{R}^{m\times n}$ ist ein Zahlenschema mit 
$m$ Zeilen und $n$ Spalten. 
- Die Matrixelemente bezeichen wir mit $a_{ij}$.
- Die Matrixmultiplikation $\mathbf{C} = \mathbf{A}\cdot \mathbf{B}$ ist definiert für   Matrizen $\mathbf{A}\in \mathbb{R}^{m\times n}, \mathbf{B}\in\mathbb{R}^{n\times k}$ und ergibt eine Matrix $\mathbf{C}\in \mathbb{R}^{m\times k}$.
- Eine allgemeine lineare Abbildung 
$$ f:\mathbb{R}^n \rightarrow \mathbb{R}^m$$ 
kann mittels einer Matrix $\mathbf{A}\in \mathbb{R}^{m\times n}$ wie folgt geschrieben werden
$$ f: \vec{x}\in \mathbb{R}^n \rightarrow \vec{y}  = \mathbf{A} \vec{x}$$.
- Tensoren sind (im Vergleich zu den zweidimensionalen Matrizen) multidimensionale Zahlenarrays (im weiteren Verlauf der Vorlesung werden uns dreistufige und vierstufige Tensoren mit 3 bzw. 4 Indizes, d.h. $a_{ijk}$ bzw $a_{ijkl}$ begegnen). 
- Für eine quadratische Matrix $\mathbf{A}\in\mathbb{R}^{n\times n}$ ist ein Eigenvektor $\vec{x}\neq \vec{0}$ ein Vektor, der unter der Abbildung $\mathbf{A}$ auf ein Vielfaches von sich selbst abgebildet wird
$$
\mathbf{A}\vec{x} = \lambda \vec{x}
$$
Symmetrische Matrizen (d.h. $\mathbf{A} = \mathbf{A}^T$ haben (a) reelle Eigenwerte ($\lambda_i \in \mathbb{R}$)
und (b) orthonormale Eigenvektoren ( $\vec{x}_i \cdot \vec{x}_j = \delta_{ij}$ ).


### Analysis und mehrdimensionale Analysis

- Die Kettenregel bestimmt die Ableitung von Funktionen des Typs $f(g(x))$. Die Ableitung ist gegeben durch
$$
f'(g(x))\cdot g'(x)
$$

---

- Eine skalare Funktion $f(x_1,\ldots,x_n)$ ist eine Abbildung $$f:\mathbb{R}^n\rightarrow \mathbb{R},$$ die einem Vektor $\vec{x}\in\mathbb{R}^n$ eine Zahl (Skalar) $f$ zuordnet. Für $n=2$ kann die Funktion als Höhenline veranschaulicht werden. Als Beispiel betrachten wir die Funktion $f(x,y) = (x-1)^2 + (y+2)^2$.

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.linspace(-3,3,100),np.linspace(-3,3,100))
z = (x-1)**2 + (y+2)**2
plt.figure()
plt.pcolormesh(x, y, z)
plt.colorbar()
plt.grid()
plt.contour(x, y, z, np.linspace(0,25,20), colors="k")
plt.gca().set_aspect("equal")
plt.show()
```

Die partielle Ableitung $\frac{\partial f}{\partial x_j}$ gibt die Änderungsrate der Funktion bezüglich der Variablen $x_j$ a. Wir berechnen die partiellen Ableitungen für obige Funktion
$$
\frac{\partial f}{\partial x} = 2(x-1)\\
\frac{\partial f}{\partial x} = 2(y+1)
$$
Der Gradient ist 
$$
\vec{\nabla} f(x,y) = \begin{pmatrix} 
\frac{\partial f}{\partial x} \\
\frac{\partial f}{\partial x} 
 \end{pmatrix}
$$
ein Vektor bestehend aus den partiellen Ableitungen. Er hat folgende Eigenschaften
- Er zeigt lokal in Richtung des steilsten Anstiegs von $f$.
- Er steht senkrecht auf den Höhenlinien von $f$.

### Aktivierungsfunktionen

Künstliche Neuronen werden über sogenannte Aktivierungsfunktionen $\sigma$ gesteuert/aktiviert. Die wichtiste Aktivierungsfunktionen sind
$$
\begin{align}
\sigma_1(x) &=& \frac{1}{e^{-x}+1}  \\
\sigma_2(x) &=& \tanh(x) \\
\sigma_3(x) &=& \max(0,x)
\end{align}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x  = np.linspace(-4,4,100)
y1 = 1.0/(np.exp(-x)+1)
y2 = np.tanh(x)
y3 = np.maximum(0,x)

plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.grid()
plt.legend(['sigma_1','sigma_2','sigma_3'])
plt.axis([-3,3,-1.5,1.5])
plt.show()
```

### Statistik

- Der Mittelwert einer statistischen Größe $x$ (d.h. es liegen Werte $x_i, i=1,\ldots,n$ vor) berechnet sie gemäß
$$
\mu = \bar{x} = \frac{1}{n}\sum_{i=1}^n x_i
$$
- Die Varianz ist die mittlere quadratische Abweichung (die Streung) vom Mittelwert. Sie errechnet sich gemäß
$$
\sigma^2 = \frac{1}{n}\sum_{i=1}^n \left(x_i-\mu\right)^2
$$
- Die sogenannte Kovarianzmatrix $\mathbf{C}$ mehrerer statistischer Größen $x^1,x^2,\ldots,x^m$ (für alle liegen die Werte $x^k_i, i=1,\ldots,n$ und $k=1,\ldots,m$ vor)berechnet sich gemäß 
$$
c_{lk} = \frac{1}{n}\sum_{i=1}^n\left(x^{l}_i-\mu_{x^l}\right)\left(x^{k}_i-\mu_{x^k}\right) = c_{kl}
$$
Die Kovarianz stellt ein Maß für den linearen Zusammenhang von Variablen dar. Ein positives Vorzeichen gibt an, dass sich die Variablen in dieselbe Richtung bewegen. Ein negatives Vorzeichen besagt das Gegenteil über den Zusammenhang aus. Ein Wert von bzw. Nahe Null deutet darauf hin, dass kein Zusammenhang besteht.
Die Kovarianzmatrix ist symmetrisch, d.h. $c_{lk}=c_{kl}$ bzw.
$\mathbf{C}=\mathbf{C}^T$. Die Eigenwerte der Kovarianzmatrix sind daher
stets reell und die Eigenvektoren bilden eine Orthonormalbasis (d.h. die Eigenvektoren sind paarweise senkrecht zueinander und haben die Länge 1).

### Faltung

Die diskrete Faltung zweier eindimensionaler arrays a und b ist definiert als 
$$ a\ast b = (a\ast b)(n) =  \sum_m a_m b_{n-m}\,.  $$ 
Anschaulich ist $(a\ast b)(n)$ der gewichtete Mittelwert von a, wobei die Gewichtung durch b gegeben ist. Das folgende Beispiel zeigt die Faltung eines verrauschten Signals $s$ mit einer Gaussfunktion $g$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

g = scipy.signal.gaussian(10,1)
g/=np.sum(g)
s = np.sin(np.linspace(0,10,200)*3) + 0.2*np.random.randn(200)
plt.grid()
s_s = np.convolve(s,g,mode='same')
plt.plot(s)
plt.plot(s_s,'r',linewidth=2)
```
